Menurut World Health Organization(WHO) standar kualitas udara adalah sebagai berikut:                                                                                                                                                                          
                                                                                             
- PM2.5: 10 μg/m3 rata-rata tahunan, dan 25 μg/m3 rata-rata dalam 24 jam.
- PM10: 20 μg/m3 rata-rata tahunan, dan 50 μg/m3 rata-rata dalam 24 jam.
- Ozone (O3): 100 μg/m3 rata-rata periode 8 jam.
- Nitrogen dioxide (NO3): 40 μg/m3 rata-rata tahunan, dan 200 μg/m3 rata-rata dalam 1 jam.
- Sulphur dioxide (SO2): 20 μg/m3 rata-rata periode 24 jam, dan 500 μg/m3 rata-rata dalam 10 jam.

```python
#merge dataset Huairou_df dan Nongzhanguan_df
Huairou_Nongzhanguan_df= pd.merge(
    left=Huairou_df,
    right=Nongzhanguan_df,
    how="outer"
)
Huairou_Nongzhanguan_df
```

```python
#merge dataset Shunyi_df dan Tiantan_df
Shunyi_Tiantan_df = pd.merge(
    left=Shunyi_df,
    right=Tiantan_df,
    how="outer"
)
Shunyi_Tiantan_df
```

```python
#merge dataset Guanyuan_Gucheng_df dengan Huairou_Nongzhanguan_df
second_one_third_df = pd.merge(
    left=Guanyuan_Gucheng_df,
    right=Huairou_Nongzhanguan_df,
    how="outer"
)
second_one_third_df
```

Setelah melakukan ekspolari data, maka dilakukan visualisasi dan eksplanatori analisis yang bertujuan untuk menampilkan data dalam bentuk visual serta menjawab pertanyaan bisnis yang telah diajukan sebelumnya.

### Pertanyaan 1: Pada tahun berapa yang memiliki rata-rata tingkat pm 10, PM2.5, dan NO2 tertinggi di seluruh daerah?

Pertama kita akan melihat terlebih dahulu sebaran data PM10, PM2.5, dan NO2 per tahun

```python
year_highest_Polutant_df = all_city_df.groupby(by=["year"]).agg({
    "PM10":"mean",
    "PM2.5":"mean",
    "NO2":"mean"
}).sort_values(by="year", ascending=True).reset_index()
year_highest_Polutant_df
```

Setelah itu membuat visualisasi untuk persebaran data rata-rata tingkat PM10, PM2.5, dan NO2 per tahun.

```python
#menampilkan grafik PM10, PM2.5, dan NO2 per tahun

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))
 
sns.barplot(y="PM10", x="year", data=year_highest_Polutant_df.sort_values(by="PM10", ascending=False), color="#72BCD4", ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("PM10", loc="center", fontsize=18)
ax[0].tick_params(axis ='both', labelsize=12)
 
sns.barplot(y="PM2.5", x="year", data=year_highest_Polutant_df.sort_values(by="PM2.5", ascending=False), color="#72BCD4", ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("PM2.5", loc="center", fontsize=18)
ax[1].tick_params(axis='both', labelsize=12)
 
sns.barplot(y="NO2", x="year", data=year_highest_Polutant_df.sort_values(by="NO2", ascending=False), color="#72BCD4", ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("NO2", loc="center", fontsize=18)
ax[2].tick_params(axis='both', labelsize=12)
 
plt.suptitle("Tingkat rata-rata polutan per tahun ", fontsize=20)
plt.show()
```

Berdasarkan visualisasi yang ditampilkan, tahun 2014 merupakan tahun dengan rata-rata tingkat PM10 tertinggi, kondisi rata-rata tingkat PM2.5 tertinggi terjadi pada tahun 2017, dan kondisi rata-rata tingkat NO2 tertinggi terjadi pada tahun 2017.

```python
#menyimpan data seluruh kota di tahun 2014 kedalam variabel air_in2014_df
air_in2014_df = all_city_df[all_city_df["year"] == 2014]
air_in2014_df.head()
```

lalu kita ambil data rata-rata PM10 berdasarkan daerah pada tahun 2014

```python
highest_2014pm10_df = air_in2014_df.groupby(by=["station","year"]).PM10.mean().sort_values(ascending=False).reset_index()
highest_2014pm10_df
```

Setelah itu kita ambil data rata-rata PM2.5 berdasarkan daerahnya.

```python
highest_2014pm25_df = air_in2014_df.groupby(by=["station","year"])["PM2.5"].mean().sort_values(ascending=False).reset_index()
highest_2014pm25_df
```

Kita juga mengambil data rata-rata NO2 berdasarkan daerahnya

```python
highest_2014NO2_df = air_in2014_df.groupby(by=["station","year"]).NO2.mean().sort_values(ascending=False).reset_index()
highest_2014NO2_df
```

```python
#menampilkan grafik PM10, PM2.5, dan NO2 pada tahun 2014


fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(20, 17))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(y="PM10", x="station", data=highest_2014pm10_df.sort_values(by="PM10", ascending=False), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("PM10", loc="center", fontsize=18)
ax[0].tick_params(axis ='both', labelsize=12)
 
sns.barplot(y="PM2.5", x="station", data=highest_2014pm25_df.sort_values(by="PM2.5", ascending=False), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("PM2.5", loc="center", fontsize=18)
ax[1].tick_params(axis='both', labelsize=12)
 
sns.barplot(y="NO2", x="station", data=highest_2014NO2_df.sort_values(by="NO2", ascending=False), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("NO2", loc="center", fontsize=18)
ax[2].tick_params(axis='both', labelsize=12)
 
plt.suptitle("Tingkat rata-rata polutan tahun 2014", fontsize=20)
plt.show()
```

### Pertanyaan 7: Pada bulan apa saja yang memiliki rata-rata tingkat PM10 tertinggi,  rata-rata tingkat PM2.5 tertinggi dan  rata-rata tingkat NO2 tertinggi pada tahun 2014?

Pertama mencari informasi rata-rata PM10, PM2.5 dan NO2 dari seluruh daerah pada tahun 2014 dengan membuat dataframe baru polutantmonth2014_df.

```python
#data berdasarkan rata2 tingkat PM10 dari yang terbesar ke terkecil
polutantmonth2014_df = all_city_df[all_city_df["year"] == 2014].resample(rule='M', on='date').agg({
    "PM10":"mean",
    "PM2.5":"mean",
    "NO2":"mean"
})
polutantmonth2014_df.index = polutantmonth2014_df.index.strftime('%B') #mengubah format order date menjadi nama bulan
 
polutantmonth2014_df = polutantmonth2014_df.reset_index()
polutantmonth2014_df
```

setelah itu kita sajikan data yang telah kita cari kita tampilkan dalam bentuk linechart

```python
plt.figure(figsize=(20, 6)) 
plt.plot(polutantmonth2014_df["date"], polutantmonth2014_df["PM10"], marker='o', linewidth=2, color="#72BCD4",label="PM10")
plt.plot(polutantmonth2014_df["date"], polutantmonth2014_df["PM2.5"], marker='o', linewidth=2, color="red", label="PM2.5")
plt.plot(polutantmonth2014_df["date"], polutantmonth2014_df["NO2"], marker='o', linewidth=2, color="orange", label="NO2")
#plt.locator_params(integer=True)
plt.title("Rata-rata tingkat polutan tahun 2014 per bulan (seluruh daerah)", loc="center", fontsize=20) 
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15) 
# show legend
plt.legend()

plt.show()
```

Berdasarkan hasil visualisasi data, diketahui bahwa:
- Untuk rata-rata PM10 mengalami puncak pada bulan Februari sebesar 156.681982, mengalami penurunan paling rendah pada bulan Juni dengan nilai rata-ratanya 74.037310, dan memuncak kembali pada bulan Oktober dengan nilai rata-rata 146.505544. Secara keseluruhan Februari menjadi bulan dengan rata-rata tingkat PM10 tertinggi pada tahun 2014.
- Untuk rata-rata PM2.5 di tahun 2014 mengalami puncak pada bulan Februari dengan niai rata-rata PM2.5 sebesar 153.808780, kemudian mengalami penurunan paling rendah  di bulan Juni dengan rata-rata PM2.5 sebesar 54.435571. Secara keseluruhan Februari menjadi bulan dengan rata-rata tingkat PM2.5 tertinggi pada tahun 2014.
- Untuk rata-rata NO2 pada tahun 2014, bulan Februari menjadi bulan dengan rata-rata NO2 tertinggi degan nilai 71.997375 dan disusul oleh bulan Oktober dengan nilai rata-rata NO2 sebesar 70.265793.
